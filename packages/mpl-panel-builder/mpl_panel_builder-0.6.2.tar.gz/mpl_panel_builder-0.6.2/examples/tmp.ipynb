```python
mpb_config = {
    "panel": {
        "dimensions": {"width_cm": 6.0, "height_cm": 5.0},
        "margins": {
            "left_cm": 1.5,
            "right_cm": 0.5,
            "top_cm": 0.5,
            "bottom_cm": 1,
        },
        "axes_separation": {
            "x_cm": 0.5,
            "y_cm": 0.5,
        },
    },
    "style": {
        "theme": "article",
        "rc_params": {
            "axes.facecolor": "none",
        },
    },
    "output": {"format": "png"},
}
```

```python
mpb.configure(mpb_config)
mpb.set_rc_style()
# fig, axs = mpb.create_panel(2, 2)
fig, axs = mpb.create_stacked_panel(2, 2)
for i in range(2):
    for j in range(2):
        axs[i][j].set(xticks=[], yticks=[])
mpb.features.draw_gridlines(fig)
```
