# Gamma


To illustrate the effect of gamma we will create a step wedge, an image with vertical bars with grey values of 0, 0.1, ... 1, and display it.

```python
wedge = np.tile( np.linspace(0, 1, 11), (11,1) )
plt.figure(figsize=(6,2))
plt.imshow(wedge, cmap='gray', aspect='auto')
ax = plt.gca()
```

and we see on the screen that the displayed intensities are not very linear.  In particular the first two bars are almost indistinguishable.  What you are seeing is the effect of the gamma encoding in the display, a nonlinear transfer function between what you send to the display and what the display shows.

If we invert the gamma encoding $I_d = I^\gamma$ by raising the grey values to the power of $1/\gamma$ then the step wedge appears much more linear.

```python
plt.figure(figsize=(6,2))
plt.imshow(wedge**(1/2.2), cmap='gray', aspect='auto')
ax = plt.gca()
```

When you load an image into an array, the values in the array are *not proportional* to the brightness in the world.  They are gamma encoded, the real-world brightness raised to the power $1/\gamma$ where $\gamma$ is around 2.2.  This can cause surprises when you compute ratios of greyscale values.
