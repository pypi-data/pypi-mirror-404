# Tutorial

1. Install using `pip install aspen-pysys`
2. Before you can use HYSYS, ensure that the simulation case you wish to work on is already open.
3. Once your simulation file is open, enter the following lines of code to set the package up.
4. An indicator that you have successfully connected with your simulation file is the printing of 'Aspen HYSYS' in the shell (or in a notebook cell if you are using a Jupyter cell or something similar).

```python
import aspen_pysys as ap
from aspen_pysys import *

ap.setup()
```

4. You can now use the aspen-pysys package in your code! To get started, refer to the following examples:

```python
feed: MaterialStream = Hysys.get_module("F")
print(feed.attr("Temperature"))
```

```python
valve1: UnitOperation = Hysys.get_module("V1")
print(valve1.attr("Pressure Drop"))
```

The package allows you to obtain a Pythonic representation of an existing HYSYS module (essentially anything you see on your flowsheet) using `Hysys.get_module(name)`, where `name` is its module.

You can then access its properties by calling `.attr(prop)`, where `prop` is a string of text containing the property you would like to access. If you're not sure about what properties you can access for a given module, just call `.attrs()` on your HYSYS module object to view all possible properties you can access.

Please also note that some properties may not exist or throw errors as they either cannot be computed or are missing values. Happy coding!

```python
separator: UnitOperation = Hysys.get_module("Separator")
separator.attrs()
```

```python
attached_products: HysysDict = separator.attr("Attached Products")
attached_products.keys()
```

```python
overhead_stream: MaterialStream = attached_products.get("V")
```
