```python
from sofascore_scraper import SofaScoreScraper

scraper = SofaScoreScraper(headless=True, verbose=True)

target_dates = ["11/04/25"]
df = scraper.scrape_matches(
    url="https://www.sofascore.com/football/tournament/spain/laliga/8",
    season="24/25",
    max_rounds=38,
    target_dates=target_dates
)
```
